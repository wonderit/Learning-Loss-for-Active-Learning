```python
import matplotlib.pyplot as plt
import csv
import pandas as pd
import numpy as np
```

```python
result = pd.read_csv('csv/k200_m0.1_s20k_tbr_l.csv', sep='\t')

# gca stands for 'get current axis'
ax = plt.gca()

k = 200
result['x'] = list(range(k, 10 * k + 1, k))

result.plot(kind='line', use_index=True, x='x', y='0.1', color='red', ax=ax)
result.plot(kind='line', use_index=True, x='x', y='0.5', color='black', ax=ax)
result.plot(kind='line', use_index=True, x='x', y='0.9', color='blue', ax=ax)
result.plot(kind='line', use_index=True, x='x', y='1', color='green', ax=ax)

plt.xlabel('Number of labeled images')
plt.ylabel('R-squared')

plt.xticks(np.arange(k, 10 * k + 1, step=k))

plt.savefig('fig/result-k200_m0.1_s20k_tbr_l2.csv.png', dpi=300)
plt.show()
```

```python
result = pd.read_csv('csv/k200_tbr_tor.csv', sep='\t')

# gca stands for 'get current axis'
ax = plt.gca()

k = 200
result['x'] = list(range(k, 10 * k + 1, k))

result.plot(kind='line', use_index=True, x='x', y='tor_0.9', color='red', ax=ax)
result.plot(kind='line', use_index=True, x='x', y='tbr_0.9', color='gray', ax=ax)
result.plot(kind='line', use_index=True, x='x', y='tor_tbr_0.9', color='blue', ax=ax)
result.plot(kind='line', use_index=True, x='x', y='tor_tbr_1.0', color='green', ax=ax)
result.plot(kind='line', use_index=True, x='x', y='random-init', color='black', ax=ax)

plt.xlabel('Number of labeled images')
plt.ylabel('R-squared')

plt.xticks(np.arange(k, 10 * k + 1, step=k))

plt.savefig('fig/result-k200-tbr-tor_all.png', dpi=300)
plt.show()
```
