```python
import matplotlib.pyplot as plt
import csv
import pandas as pd
import numpy as np

result = pd.read_csv('csv/k200_subset_random.csv', sep='\t')

# gca stands for 'get current axis'
ax = plt.gca()

k = 200
result['x'] = list(range(k, 10 * k + 1, k))

result.plot(kind='line', use_index=True, x='x', y='2k', color='red', ax=ax)
result.plot(kind='line', use_index=True, x='x', y='10k', color='black', ax=ax)
result.plot(kind='line', use_index=True, x='x', y='20k', color='blue', ax=ax)

plt.xlabel('Number of labeled images')
plt.ylabel('R-squared')

plt.xticks(np.arange(k, 10 * k + 1, step=k))

plt.savefig('fig/result-k200-subset-ll4al.png', dpi=300)
plt.show()
```
